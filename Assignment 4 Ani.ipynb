## 4) COMPUTATION: Finite Difference Time Domain method

```python
x_space = np.arange(-200, 200, 1)

def E_field(t, x=x_space, E=1, s=10, t0=0):
    E_field = E * exp(-(x - (t - t0))**2 / s**2)
    return E_field
```

```python
t_max = 250
imp0 = 377
# H_y = np.zeros((np.shape(x_space)[0], t_max))
# E_z = np.zeros((np.shape(x_space)[0], t_max))

# for i in range(0, t_max, 1):
#     for j in range(1, len(x_space), 1):
#         H_y[j-1, :] += (E_z[j, :] - E_z[j-1, :]) / imp0
#         E_z[j, :] += (H_y[j, :] - H_y[j-1, :]) * imp0
#         j += 1
# #    E_z[:, 0] = E_field(x=x_space, t=i)
#     E_z[:, 0] = exp(-(i-30)*(i-30)/100)
#     i += 1

# Direct copy of code
# ===================
H_y = np.zeros((t_max, len(x_space)))
E_z = np.zeros((t_max, len(x_space)))
temp1 = np.zeros(len(x_space))
temp2 = np.zeros(len(x_space))


for i in range(0, t_max, 1):
    temp1[len(x_space)-1] = temp1[len(x_space)-2]
    for j in range(0, len(x_space)-1, 1):
        temp1[j] += (temp2[j+1] - temp2[j]) / imp0
    
    temp2[0] = temp2[1]
    for j in range(1, len(x_space), 1):    
        temp2[j] += (temp1[j] - temp1[j-1]) * imp0
    
    temp2[200] = exp(-(i-30)*(i-30)/100)
    
    E_z[i,:] = temp2[:]
    H_y[i,:] = temp1[:]
    
# for i in range(0, t_max, 1):
#     for j in range(1, len(x_space)-1, 1):
#         H_y[j-1] = (E_z[j] - E_z[j-1]) / imp0
#         E_z[j] = (H_y[j] - H_y[j-1]) * imp0
#         j += 1
#     E_z[50] += exp(-(i-30)*(i-30)/100)
#     i += 1
```

```python
plt.figure(figsize = (10, 20))
for i in range(0, 250, 5):
#     plt.figure(figsize=(10, 4))
    plt.plot(x_space, E_z[i, :] + i/5, color="b", zorder=1)
    plt.plot(x_space, H_y[i, :]*100 + i/5, color="orange", zorder=0)
#     plt.ylim(-2, 2)
    i+=1
# np.shape(E_z)
plt.legend(('E', 'B'))
plt.show()
```
